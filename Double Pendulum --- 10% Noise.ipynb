```python
def double_pendulum(X):
    
    g,m1,m2,l1,l2 = (10,1,1,1,1)
    t1, t2, p1, p2 = X
    
    C1 = p1*p2*np.sin(t1-t2)/(l1*l2*(m1+m2*np.sin(t1-t2)**2))
    C2 = (l2**2*m2*p1**2+l1**2*(m1+m2)*p2**2-l1*l2*m2*p1*p2*np.cos(t1-t2))*np.sin(2*(t1-t2))/(2*(l1*l2*(m1+m2*np.sin(t1-t2)**2))**2)

    d_t1 = (l2*p1-l1*p2*np.cos(t1-t2))/(l1**2*l2*(m1+m2*np.sin(t1-t2)**2))
    d_t2 = (l1*(m1+m2)*p2-l2*m2*p1*np.cos(t1-t2))/(l1*l2**2*m2*(m1+m2*np.sin(t1-t2)**2))
    d_p1 = -(m1+m2)*g*l1*np.sin(t1)-C1+C2
    d_p2 = -m2*g*l2*np.sin(t2)+C1-C2

    return [d_t1, d_t2, d_p1, d_p2]

true_vector_field = np.stack(double_pendulum(X), axis = 0)
approx_vec_field = sess.run(simple_net(placeholders['Y_0'], weights, biases), feed_dict = {placeholders['Y_0']: X})

noise_error = np.linalg.norm(measurement_noise - N.eval(),'fro')**2 / m
vector_field_error = np.linalg.norm(true_vector_field - approx_vec_field,'fro')**2 / np.linalg.norm(true_vector_field,'fro')**2

print(noise_error)
print(vector_field_error)
```
